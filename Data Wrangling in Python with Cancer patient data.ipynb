```python
# Data Wrangling with Cancer Patient Data
# By Symon Kimitei
#============================================================

# Load libraries and other dependencies as necessary
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
%matplotlib inline

# Set working directory
os.chdir('C:\Users\skimitei\Desktop\python\cancerpatient')
os.getcwd()

```

```python
# Load data set and subset the data set to contain only the patients who have died
data = pd.read_csv('data_patient.txt', delimiter='\t')
df = pd.DataFrame(data, columns = ['AGE', 'RACE','OS_MONTHS','VITAL_STATUS'])
```

```python
# Make a histogram that shows a distribution of those who have died from Cancer by age
#----------------------------------------------------------------------------------------
# First, subset the data to extract only the list of patients who have died
newdf = df[df['VITAL_STATUS'] == 'Dead']

# Generate the histograms
newdf.hist('AGE')
plt.title('A histogram of the dead by Age distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Calculate the average age of those who have died from cancer and display
print('Average age is:', sum(newdf['AGE'])/len(newdf['AGE']))
```

```python
# Make a scatter plot
plt.plot( 'AGE', 'OS_MONTHS', data=newdf, linestyle='none', marker='o', alpha=0.8)
plt.title('A scatter plot showing the correlation between Age & Os_months')
plt.xlabel('Age')
plt.ylabel('Os_months')
x = np.linspace(0, 10, 100)
plt.yscale('log')
plt.show()
```

```python
#Create a box plot
# First split the data and then create a dictionary out of it. Finally create a side by side box plot
under50 = newdf[newdf['AGE']<=50]
over50 = newdf[newdf['AGE']>50]

# Create a dictionary
my_dict = {'Under 50 years': under50['AGE'], 'Over 50 years': over50['AGE']}
fig, ax = plt.subplots()
ax.boxplot(my_dict.values())
ax.set_xticklabels(my_dict.keys())
```
